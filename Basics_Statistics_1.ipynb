```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the datset
df = pd.read_csv('C:/Users/victus/Data_Science/New folder/Basic stats - 1/sales_data_with_discounts.csv')
df
```

```python
#Get the numerical columns  
numerical_columns = df.select_dtypes(include=[np.number]).columns.tolist()
numerical_columns 
```

```python
df[numerical_columns]
```

```python
#Mean of numerical columns
df[numerical_columns].mean()
```

```python
#Median of numerical columns
df[numerical_columns].median() 
```

```python
#Mode of numerical columns
df[numerical_columns].mode()  
```

```python
#Standard deviation of numerical columns
df[numerical_columns].std()  
```

```python
#Variance of numerical columns
df[numerical_columns].var() 
```

```python
#Numerical columns
df.describe(include=[np.number])
```

```python
#Count of missing values in numerical columns
df[numerical_columns].isnull().sum()
```

```python
#Skewness of numerical columns
df[numerical_columns].skew()
```

```python
#Correlation matrix of numerical columns
df[numerical_columns].corr()
```

```python
#Histogram of numerical columns
for col in numerical_columns:
    plt.figure(figsize=(10, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#Boxplots for numerical columns
for col in numerical_columns:
    plt.figure(figsize=(10,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.show()
```

```python
#Get the categorical columns
categorical_columns = df.select_dtypes(include=[object]).columns.tolist()
categorical_columns
```

```python
#Count of missing values in categorical columns
df[categorical_columns].isnull().sum()
```

```python
for col in categorical_columns:
    plt.figure(figsize=(10,4))
    sns.countplot(data=df, x=col,order=df[col].value_counts().index)
    plt.title(f'Bar Chart of {col}')
    plt.xlabel('col')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
df_standardized = df.copy()
df_standardized[numerical_columns] = StandardScaler().fit_transform(df[numerical_columns])

# Compare before and after
print("Before Standardization:")
print(df[numerical_columns].head())

print("\nAfter Standardization:")
print(df_standardized[numerical_columns].head())
```

```python
# One-Hot Encoding for categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

# Show BEFORE transformation
print("Before One-Hot Encoding:")
print(df[categorical_columns].head())

# Show AFTER transformation
print("\nAfter One-Hot Encoding:")
print(df_encoded.head())
```
